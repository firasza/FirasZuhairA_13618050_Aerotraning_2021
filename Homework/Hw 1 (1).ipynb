```python
i=1
while i <= 100:
    if i%15==0:
        print("FizzBuzz", end="")
    elif i%5==0:
        print("Buzz", end="")
    elif i%3==0:
        print("Fizz",end="")
    else:
        print(i, end="")
    print()
    i+=1
```

Pada kode ini akan dilakukan looping hingga angka 100. Dengan melakukan pergantian angka menjadi kata meggunakan if
-Melakukan cek apakah angka tersebut habis dibagi 15. Jika habis, maka output yang dikeluarkan berupa "FizzBuzz"
-Melakukan cek apakah angka tersebut habis dibagi 5. Jika habis, maka output yang dikeluarkan berupa "Buzz"
-Melakukan cek apakah angka tersebut habis dibagi 3. Jika habis, maka output yang dikeluarkan berupa "Fizz"
-Selain kondisi diatas, akan keluar output berupa angka
